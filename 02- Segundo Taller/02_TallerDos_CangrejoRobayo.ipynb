### 2. Crear el Dataset ✨
Ejecuta la siguiente celda para crear los DataFrames que usaremos en los ejercicios.

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración de estilo
sns.set_theme(style='whitegrid')
np.random.seed(42)

# Dataset 1: Ventas mensuales
meses = ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic']
ventas_2022 = [45, 52, 48, 65, 70, 85, 90, 88, 75, 80, 95, 110]
ventas_2023 = [50, 58, 55, 72, 78, 92, 98, 95, 82, 88, 105, 120]

df_ventas = pd.DataFrame({
    'Mes': meses,
    'Ventas_2022': ventas_2022,
    'Ventas_2023': ventas_2023
})

# Dataset 2: Datos de empleados
n_empleados = 150
df_empleados = pd.DataFrame({
    'Edad': np.random.normal(35, 8, n_empleados).astype(int).clip(22, 65),
    'Salario': np.random.normal(50000, 15000, n_empleados).clip(25000, 120000),
    'Años_Experiencia': np.random.normal(8, 4, n_empleados).clip(0, 30),
    'Departamento': np.random.choice(['Ventas', 'IT', 'Marketing', 'RRHH'], n_empleados),
    'Satisfaccion': np.random.randint(1, 11, n_empleados)
})

# Dataset 3: Productos
productos = ['Laptop', 'Mouse', 'Teclado', 'Monitor', 'Auriculares']
unidades_vendidas = [120, 450, 380, 210, 290]

df_productos = pd.DataFrame({
    'Producto': productos,
    'Unidades': unidades_vendidas
})

print('Datasets creados exitosamente')
print(f'\ndf_ventas: {df_ventas.shape}')
print(f'df_empleados: {df_empleados.shape}')
print(f'df_productos: {df_productos.shape}')
````

### 3. Gráfico de Líneas Simple ✨
Crea un gráfico de líneas que muestre las ventas de 2022 (`Ventas_2022`) a lo largo de los meses. Usa `plt.plot()` y añade título y etiquetas a los ejes.

````python
plt.figure(figsize=(10,6)) # Establece el tamaño de la figura (ancho, alto) en pulgadas
plt.plot(df_ventas['Mes'], df_ventas['Ventas_2022']) # Crea el gráfico de líneas usando 'Mes' para el eje X y 'Ventas_2022' para el eje Y
plt.title('Ventas Mensuales 2022') # Establece el título del gráfico
plt.xlabel('Meses') # Etiqueta del eje X
plt.ylabel('Ventas') # Etiqueta del eje Y
plt.xticks(rotation=45) # Rota las etiquetas del eje X 45 grados para mejorar la legibilidad
plt.grid(True) # Añade una cuadrícula al gráfico
plt.tight_layout() # Ajusta automáticamente los parámetros de la subtrama para un diseño ajustado
plt.show() # Muestra el gráfico en pantalla
# plt.savefig('mi grafico3.png') # Línea comentada para guardar el gráfico si se desea, se puede descomentar para usarla
````

### 15. Heatmap de Correlación ✨
Crea un heatmap que muestre la matriz de correlación de las variables numéricas de `df_empleados`. Usa `sns.heatmap()` con `annot=True`.

### 17. Subplots (Múltiples Gráficos) ✨
Crea una figura con 2 subplots (1 fila, 2 columnas):
- Subplot 1: Histograma de edades
- Subplot 2: Histograma de salarios

Usa `plt.subplots()`.

````python
fig, axes = plt.subplots(2, 2, figsize=(15, 10)) # Crea una figura con una cuadrícula de 2 filas y 2 columnas para los subplots

axes[0, 0].hist(df_empleados['Edad'], bins=15) # Primer subplot (arriba a la izquierda): Histograma de Edades
axes[0, 0].set_title('Distribución de Edades (Superior Izquierda)')
axes[0, 0].set_xlabel('Edad')
axes[0, 0].set_ylabel('Frecuencia')

axes[0, 1].hist(df_empleados['Salario'], bins=15, color='orange') # Segundo subplot (arriba a la derecha): Histograma de Salarios
axes[0, 1].set_title('Distribución de Salarios (Superior Derecha)')
axes[0, 1].set_xlabel('Salario')
axes[0, 1].set_ylabel('Frecuencia')

axes[1, 0].hist(df_empleados['Edad'], bins=15, color='green') # Tercer subplot (abajo a la izquierda): Otro Histograma de Edades
axes[1, 0].set_title('Distribución de Edades (Inferior Izquierda)')
axes[1, 0].set_xlabel('Edad')
axes[1, 0].set_ylabel('Frecuencia')

axes[1, 1].hist(df_empleados['Salario'], bins=15, color='purple') # Cuarto subplot (abajo a la derecha): Otro Histograma de Salarios
axes[1, 1].set_title('Distribución de Salarios (Inferior Derecha)')
axes[1, 1].set_xlabel('Salario')
axes[1, 1].set_ylabel('Frecuencia')

plt.tight_layout() # Ajusta automáticamente los parámetros de la subtrama para un diseño ajustado y evita superposiciones
plt.show() # Muestra la figura con todos los subplots
````

##Conclusion 🍀

1. Las edades se concentran principalmente entre los 25 y 40 años en ambos histogramas, mostrando una distribución similar.

2. Los salarios se agrupan entre 40.000 y 70.000, aunque hay algunos valores altos aislados por encima de 80.000.

3. Las distribuciones superior e inferior son prácticamente iguales, mostrando consistencia en la edad y el salario de los empleados.
