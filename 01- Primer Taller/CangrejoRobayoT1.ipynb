### 3. Crear el DataFrame ‚õÖ

Ejecuta la siguiente celda para crear el DataFrame `df_ventas` que usaremos en los ejercicios.

Crea un nuevo DataFrame que contenga √∫nicamente las ventas realizadas en la ciudad de 'Bogot√°'.

### 10. Filtrar por Condici√≥n Num√©rica ‚ù£
Filtra el DataFrame para mostrar solo las filas donde la `Cantidad` vendida fue mayor a 15.

### 11. Filtrar por M√∫ltiples Condiciones ‚ù§
Muestra las ventas de 'Laptop' donde la `Cantidad` fue superior a 8.

### 13. Crear una Nueva Columna ‚õÖ
Crea una nueva columna llamada `Ingreso_Total` que sea el resultado de multiplicar `Cantidad` por `Precio_Unitario`.

```python
df_ventas["Ingreso_Total"] = df_ventas["Cantidad"] * df_ventas["Precio_Unitario"]
df_ventas

```

### 14. Manejo de Valores Nulos ‚õÖ
Rellena los valores nulos en la columna `Vendedor_ID` con el valor `0`.

```python
df_ventas ["Vendedor_ID"] = df_ventas["Vendedor_ID"].fillna(0)
df_ventas
```

### 15. Aplicar una Funci√≥n (`.apply`) ‚õÖ
Crea una columna `Categoria_Precio` que contenga 'Alto' si el `Precio_Unitario` es mayor a 500, y 'Bajo' en caso contrario. Usa una funci√≥n y `.apply()`.

```python
def clasificar_precio(valor):
    if valor > 500:
        return "Alto"
    else:
        return "Bajo"

df_ventas["Categoria_Precio"] = df_ventas["Precio_Unitario"].apply(clasificar_precio)
df_ventas
```

### 16. Ordenar el DataFrame ‚õÖ
Ordena `df_ventas` por la columna `Ingreso_Total` de forma descendente.

```python
df_ventas_ordenado = df_ventas.sort_values(by="Ingreso_Total", ascending=False)
df_ventas_ordenado

```

### 17. Agrupaci√≥n Simple (`.groupby`) ‚õÖ
Agrupa el DataFrame por `Ciudad` y calcula la suma de `Ingreso_Total` para cada ciudad.

```python
df_agrupado_ciudad = df_ventas.groupby("Ciudad")["Ingreso_Total"].sum()
df_agrupado_ciudad
```

### 18. Agrupaci√≥n con M√∫ltiples Agregaciones ‚õÖ
Agrupa por `Producto` y calcula la cantidad promedio (`mean`) vendida y el ingreso total (`sum`) para cada producto.

```python
df_agrupado_producto = df_ventas.groupby("Producto").agg({
    "Cantidad": "mean",
    "Ingreso_Total": "sum"
})

df_agrupado_producto

```

### 19. Encontrar el Producto m√°s Vendido ‚õÖ
Encuentra el nombre del producto con el mayor `Ingreso_Total` acumulado.

```python
producto_top = df_ventas.groupby("Producto")["Ingreso_Total"].sum().idxmax()
producto_top

```
